```python
actor.shape
actor.info()
```

## Primer paso: transformar a lowercase los nombres y apellidos, y luego juntarlos en una sola columna

```python
actor['first_name'] = actor['first_name'].astype('str') 
actor['last_name'] = actor['last_name'].astype('str') 
actor.info()
```

```python
actor['first_name'] = actor['first_name'].str.lower()
actor['last_name'] = actor['last_name'].str.lower()
```

```python
actor["Full_Name"] = actor['first_name'] + " " + actor['last_name']
actor.head()
```
